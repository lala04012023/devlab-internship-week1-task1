```python
top_countries = df["COUNTRY"].value_counts().head(5)
top_countries.plot(kind="bar",color='skyblue',edgecolor='black')
plt.title("Saya əsasən TOP 5 ölkə sıralaması")
plt.xlabel("Ölkələr")
plt.ylabel("sifarişlər")
plt.show()
```

```python
plt.hist(df["SALES"], bins=20,color='yellow',edgecolor='black')

plt.title("Satış paylanması")
plt.xlabel("Satışlar")
plt.ylabel("Tezlik")

plt.show()
```
